```python
import pandas as pd
import plotly.express as px

from dash import html
from dash import dcc
import dash
from dash.dependencies import Input, Output, State
```

```python
data=pd.read_csv("dataforscatter.csv")
data.head()
```

```python
list_loc = data["location"].to_list()
list_loc = [x.split(",") for x in list_loc]
list_lat=[x[0] for x in list_loc]
list_lat=[x.split("(") for x in list_lat]
list_lat=[float(x[1]) for x in list_lat]
data["latitude"]=list_lat
```

```python
list_long=[x[1] for x in list_loc]
list_long=[x.split(")") for x in list_long]
list_long=[float(x[0]) for x in list_long]
data["longitude"]=list_long
```

```python
app = dash.Dash(__name__)


app.layout = html.Div([
        html.Div([html.Span("Map of New York City & the schools"),
            dcc.Graph(id="map", figure={}),     #id to manipulate the graph later
            dcc.Dropdown(
            id="Borough",
            options=[    
                {"label": "Bronx", "value": "Bronx"},
                {"label": "Brooklyn", "value": "Brooklyn"},
                {"label": "Manhattan", "value": "Manhattan"},
                {"label": "Queens", "value": "Queens"},
                {"label": "Staten Island", "value": "Staten Island"},
                {"label": "All", "value":"all"}
            ],
            value="all",
            searchable=False,
            style={'width': '40%', 'height': '35px'},
            placeholder="Please select area"          
                        )])
])
@app.callback(
    Output(component_id="map", component_property="figure"),
    Input(component_id="Borough", component_property="value"))
    
def update_map(borough_chosen):  #input #starts statement
    if borough_chosen=="all":
        df=data
        zoom=9.5
    else:
        df=data[(data["borough"]==borough_chosen)]
        zoom=10
    
    fig=px.scatter_mapbox(df, lat="latitude", lon="longitude", 
                    hover_name="school_name", hover_data=["cohort_year", "total_grads_of_cohort"],
                    color_discrete_sequence=["fuchsia"], zoom=zoom, height=600, width=800)
    fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    return fig  #component_property of output


if __name__ == '__main__':
    app.run_server(debug=False)
```
